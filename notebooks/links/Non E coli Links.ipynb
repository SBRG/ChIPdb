## Get the Data

```python
search_path = '../ChiPdb/data/tfs_search.json'
sa_imdb_path = '../../iModulonDB/final_im_objects/s_aureus_staph_precise_imdb.json'
pa_imdb_path = '../../iModulonDB/final_im_objects/p_aeruginosa_modulome_imdb.json'
mt_imdb_path = '../../iModulonDB/final_im_objects/m_tuberculosis_modulome_imdb.json'
link_path = 'non_e_coli_link_df.csv'

tf_info = pd.read_json(search_path)
sa_ica_data = load_json_model(sa_imdb_path)
pa_ica_data = load_json_model(pa_imdb_path)
mt_ica_data = load_json_model(mt_imdb_path)

link_df = pd.read_csv(link_path, index_col = 0)
```

## iModulons

```python
im_object = {'Pseudomonas aeruginosa': pa_ica_data,
             'Mycobacterium tuberculosis': mt_ica_data,
             'Staphylococcus aureus': sa_ica_data}
url_pre = {'Pseudomonas aeruginosa': 
           'https://imodulondb.org/iModulon.html?organism=p_aeruginosa&dataset=modulome&k=',
           'Mycobacterium tuberculosis': 
           'https://imodulondb.org/iModulon.html?organism=m_tuberculosis&dataset=modulome&k=',
           'Staphylococcus aureus': 
           'https://imodulondb.org/iModulon.html?organism=s_aureus&dataset=staph_precise&k='}

# get the regulators for each iModulon
tf_to_im_dict = {}
for o, ica_data in im_object.items():
    tf_to_im = pd.DataFrame()
    for k, regstr in zip(ica_data.imodulon_table.index, ica_data.imodulon_table.regulator):
        if not(isinstance(regstr, str)):
            continue

        for tf in regstr.replace('/','+').split('+'):
            if tf in tf_to_im.index:
                tf_to_im.loc[tf, tf_to_im.loc[tf].count()] = k
            else:
                tf_to_im.loc[tf, 0] = k
    
    # remove everything but those that have ChIP data
    link_df_o = link_df.loc[link_df.Organism == o]
    rel_tfs = tf_to_im.loc[tf_to_im.index.isin(link_df_o.TF)].dropna(1, 'all')
    
    tf_to_im_dict[o] = rel_tfs

# make links
def im_link(organism, im_name):
    ica_data = im_object[organism]
    imname2num = dict(zip(ica_data.imodulon_table.index, range(ica_data.imodulon_table.shape[0])))
    return url_pre[organism] + str(imname2num[im_name])
```

## Output

```python
# make an organism dictionary
organism = {}
for o in link_df.Organism.unique():
    genus_species = o.split(' ')
    organism[o] = genus_species[0][0].lower() +'_'+ genus_species[1]
```

```python
# output!
output_folder = '../ChiPdb/data/all_other/outlinks/'

for i, row in link_df.iterrows():
    res = row.drop(['TF','Organism','Strain']).fillna('')
    tf = row.TF
    o = row.Organism
    o_str = organism[row.Organism]
    
    if o in tf_to_im_dict.keys():
        if tf in tf_to_im_dict[o].index:
            for k in tf_to_im_dict[o].loc[tf].dropna():
                res[k] = im_link(o, k)
            print(o, tf)

    res.to_csv(output_folder + o_str + '_' + tf + '_links.csv', header = None)
```
